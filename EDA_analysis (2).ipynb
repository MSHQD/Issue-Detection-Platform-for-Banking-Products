```python
import inspect
from collections import namedtuple
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
import pymorphy2
from sklearn.feature_extraction.text import CountVectorizer

ArgSpec = namedtuple('ArgSpec', 'args varargs keywords defaults')
def getargspec(func):
    spec = inspect.getfullargspec(func)
    return ArgSpec(spec.args, spec.varargs, spec.varkw, spec.defaults)
inspect.getargspec = getargspec

nltk.download('stopwords')

morph = pymorphy2.MorphAnalyzer()
ru_stopwords = set(stopwords.words('russian'))

bad_pos = {'PREP', 'CONJ', 'PRCL', 'INTJ', 'NPRO'}

domain_stopwords = {
    'газпромбанк', 'банк', 'отзыв', 'достоинство', 'недостаток',
    'сотрудник', 'клиент', 'день', 'месяц', 'год', 'час', 'минута', 'вопрос', 'средство', 'итог', 'ответ', 'рубль', 'вопрос',
    'человек', 'телефон', 'руб', 'номер', 'ситуация'
}

def normalize_text(text):
    if pd.isna(text):
        return ''
    text = str(text).lower().replace('ё', 'е')
    text = re.sub(r'http\S+|www\S+', ' ', text)
    text = re.sub(r'<.*?>', ' ', text)
    text = re.sub(r'403 forbidden', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = re.sub(r'[^а-я\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def prepare_unigrams(text):
    text = normalize_text(text)
    lemmas = []
    for word in text.split():
        if len(word) < 3 or word in ru_stopwords:
            continue
        p = morph.parse(word)[0]
        pos = p.tag.POS
        lemma = p.normal_form

        if pos in bad_pos:
            continue
        if pos != 'NOUN':
            continue
        if lemma in domain_stopwords or lemma in ru_stopwords:
            continue

        lemmas.append(lemma)
    return ' '.join(lemmas)

def prepare_phrases(text):
    text = normalize_text(text)
    tokens = []
    parsed = []

    for word in text.split():
        if len(word) < 3 or word in ru_stopwords:
            continue
        p = morph.parse(word)[0]
        pos = p.tag.POS
        lemma = p.normal_form

        if pos in bad_pos:
            continue
        if lemma in domain_stopwords or lemma in ru_stopwords:
            continue

        parsed.append((lemma, pos))

    phrases = []
    for i in range(len(parsed) - 1):
        w1, p1 = parsed[i]
        w2, p2 = parsed[i + 1]

        if p1 in {'ADJF', 'ADJS'} and p2 == 'NOUN':
            phrases.append(f'{w1} {w2}')
        elif p1 == 'NOUN' and p2 == 'NOUN':
            phrases.append(f'{w1} {w2}')

    return ' | '.join(phrases)

df_text = df[df['text'].notna()].copy()

df_text['clean_unigrams'] = df_text['text'].apply(prepare_unigrams)

vec_words = CountVectorizer()
X_words = vec_words.fit_transform(df_text['clean_unigrams'])
word_freq = pd.DataFrame({
    'term': vec_words.get_feature_names_out(),
    'count': X_words.toarray().sum(axis=0)
}).sort_values('count', ascending=False)

display(word_freq.head(15))

df_text['clean_phrases'] = df_text['text'].apply(prepare_phrases)

all_phrases = ' | '.join(df_text['clean_phrases'])
phrase_list = [p.strip() for p in all_phrases.split('|') if p.strip()]

phrase_freq = (
    pd.Series(phrase_list)
    .value_counts()
    .reset_index()
)
phrase_freq.columns = ['term', 'count']

display(phrase_freq.head(15))
```

```python
def detect_services(text):

    if pd.isna(text):
        return []

    text = text.lower()
    found = []

    for service, keywords in services.items():
        for word in keywords:
            if word in text:
                found.append(service)
                break

    return found
```

### Частотность слов в негативных и позитивных отзывах.

```python
import matplotlib.pyplot as plt

df_text['sentiment'] = df_text['rating'].apply(
    lambda x: 'positive' if x >= 4 else ('negative' if x <= 2 else 'neutral')
)

services_exploded = df_text.explode('services')

filtered = services_exploded[services_exploded['sentiment'].isin(['positive', 'negative'])]

service_sentiment = (
    filtered.groupby(['services', 'sentiment'])
    .size()
    .unstack(fill_value=0)
)

service_sentiment['total'] = service_sentiment.sum(axis=1)
service_sentiment = service_sentiment.sort_values('total')

plt.figure(figsize=(10,6))

plt.barh(service_sentiment.index, service_sentiment['negative'], label='Негативные')
plt.barh(service_sentiment.index, service_sentiment['positive'],
         left=service_sentiment['negative'], label='Позитивные')

plt.title("Распределение отзывов по услугам (позитив / негатив)")
plt.xlabel("Количество отзывов")

plt.legend()
plt.show()
```

```python
 df_sent = df_text[df_text['source'] != 'sravni'].copy()

negative_texts = df_sent[df_sent['rating'] <= 2]['clean_unigrams']
positive_texts = df_sent[df_sent['rating'] >= 4]['clean_unigrams']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

# негативные
X_neg = vectorizer.fit_transform(negative_texts)
neg_counts = X_neg.toarray().sum(axis=0)
neg_words = pd.DataFrame({
    'word': vectorizer.get_feature_names_out(),
    'count': neg_counts
}).sort_values('count', ascending=False)


display(neg_words.head(15))
```

```python
# позитивные
X_pos = vectorizer.fit_transform(positive_texts)
pos_counts = X_pos.toarray().sum(axis=0)
pos_words = pd.DataFrame({
    'word': vectorizer.get_feature_names_out(),
    'count': pos_counts
}).sort_values('count', ascending=False)

display(pos_words.head(15))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

df_sent = df_text[df_text['rating'].notna()].copy()

negative_texts = df_sent[df_sent['rating'] <= 2]['clean_unigrams']
positive_texts = df_sent[df_sent['rating'] >= 4]['clean_unigrams']

vectorizer = CountVectorizer()
vectorizer.fit(pd.concat([negative_texts, positive_texts]))

X_neg = vectorizer.transform(negative_texts)
X_pos = vectorizer.transform(positive_texts)

neg_counts = X_neg.toarray().sum(axis=0)
pos_counts = X_pos.toarray().sum(axis=0)

compare_words = pd.DataFrame({
    'word': vectorizer.get_feature_names_out(),
    'negative_count': neg_counts,
    'positive_count': pos_counts
})

compare_words['total'] = compare_words['negative_count'] + compare_words['positive_count']
top_words = compare_words.sort_values('total', ascending=False).head(10)

y = np.arange(len(top_words))
height = 0.4

plt.figure(figsize=(10,6))

plt.barh(y - height/2, top_words['negative_count'], height, label='Негативные')
plt.barh(y + height/2, top_words['positive_count'], height, label='Позитивные')

plt.yticks(y, top_words['word'])

plt.xlabel("Количество упоминаний")
plt.title("Сравнение частоты слов (негатив vs позитив)")

plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()
```
